```
%sql
drop view if exists gold.Accounts
```

```
%sql
create view gold.Accounts
as
  select 
       a.accountid as DynamicsGuid
     , a.accountnumber
     , a.modifiedon
     , a.industrycode
     , a.BillingAccountId as AccountId
     , a.Culture
     , a.AccountName
     , a.Status
     , a.IsTest
     , a.DataPurgedCompleted
     , a.CreationDate
     , a.ModificationDate
     , a.OldAccountId
     , ca.CompanyCount
  from delta.`/mnt/silver/cab/accounts` as a
    left join ( select AccountId, count(*) as CompanyCount 
                from delta.`/mnt/silver/cab/companies`
                group by InstanceId, AccountId 
              ) ca on ca.AccountId = a.BillingAccountId 
```
