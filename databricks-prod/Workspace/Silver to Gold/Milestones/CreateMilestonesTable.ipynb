```
dbutils.widgets.combobox(TextConstant.environment, Environment.qa.name, [l.name for l in Environment])
environment = Environment[dbutils.widgets.get(TextConstant.environment)]

lakeName = environment.value['dataLakeName']
```

```
# get permissions
spark.conf.set(
  "fs.azure.account.key."+lakeName+".blob.core.windows.net",
  dbutils.secrets.get(scope='default',key="data-lake-storage-key")
)  
```

```
%sql
-- drop table if exists gold.Milestones
```

```
command = """
create table if not exists gold.Milestones (
  Id integer,
  Source string,
  Name string,
  StartDate timestamp,
  EndDate timestamp,
  Notes string,
  RelatedEventId integer  
)
USING DELTA
LOCATION 'wasbs://gold@"""+lakeName+""".blob.core.windows.net/cab/milestones'"""

spark.sql(command)
```
