```
%sql
drop view if exists gold.Messages
```

```
%sql
create view gold.Messages
as
  select 
        InstanceId
      , CompanyId
      , FromUserName
      , MessageId
      , SentTimeUtc
      , SentTimeLocal
      , DeletedChangeTimeUtc
      , Deleted
      , MessageSubject
      , MessageContent
      , ToUsers
      , ModifiedDateUtc
      , CreatedDateUtc
  from delta.`/mnt/silver/cab/sent_messages` 
```
