```
%sql
create view gold.Orders
as
  select 
        InstanceId
      , CompanyId
      , CreatedUTCTime
      , OrderId
      , StartTime
      , OrderNumberInt
      , OrderNumber
      , ClientName
      , OrderModuleId
      , ModuleName
      , OrderModuleVersion
      , CurrentStatusIndex
      , CurrentStatusName as CurrentStatus
      , Latitude
      , Longitude
      , AddressLine1
      , AddressLine2
      , City
      , RegionName
      , CountryCode
      , PostalCode
      , ContactName
      , ContactPhoneNumber
      , ModifiedUtc
      , IsClosed
      , CreatedByUser
      , Version
      , ContactEmail
      , ModifiedByUser
      , SubRegionName
      , RegionCode
      , SubRegionCode
      , CreatedUtc
      , ExternalId
      , CurrentOrderStatusId
      , CurrentUserFullName as CurrentUser
      , CurrentStatusStartLocalTime
      , CreatedLocalTime
      , CurrentStatusStartUtcTime
      , DurationInMinutes
      , EndTime
  from delta.`/mnt/silver/cab/order_headers` 
```
