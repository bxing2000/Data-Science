```
%sql
create view gold.Devices
as
  select
        InstanceId
      , DeviceId
      , DeviceType
      , DeviceTypeName
      , SerialNumber
      , IsPhysical
      , Deleted
      , CreatedUtc
      , ModifiedUtc
      , CompanyId
      , Name
      , TimeZoneId
      , DeviceTypeDisplayName
      , DeviceModelId
      , DeviceDataProviderId
      , DeviceName
      , Phone
      , Sim
      , IMEI
      , CustomerWarrantyDate
      , Active
      , ManufacturerWarrantyDate
      , ServicePlan
      , DeviceNameString
      , DeviceNameNumber
      , ActivationDate
      , DeactivationDate
  from delta.`/mnt/silver/cab/devices` 
```
