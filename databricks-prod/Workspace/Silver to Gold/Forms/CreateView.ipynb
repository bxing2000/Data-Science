```
%sql
create view gold.Forms
as
  select 
        f.InstanceId
      , f.CompanyId
      , c.CompanyName
      , f.FormHeaderId
      , f.CompanyModuleId
      , f.Modulename
      , f.CompanyModuleVersion
      , f.StartedByUserId
      , f.StartUtcTimeTag
      , f.StartLocalTimeTag
      , f.FormStatusUtcTimeTag
      , f.FormStatusLocalTimeTag
      , f.StartSubRegionName
      , f.StartLat
      , f.StartLon
      , f.StartGeofenceName
      , f.StartGPSProcessed
      , f.StartAddressLine1
      , f.StartAddressLine2
      , f.StartCity
      , f.StartRegionName
      , f.StartPostalCode
      , f.StartCountryCode
      , f.StartRegionCode
      , f.StartSubRegionCode
      , f.CreatedUtc
      , f.ModifiedUtc
      , f.Reviewed
      , f.ReviewedBy
      , f.ReviewedUtc
      , f.SharedLinkId
      , f.CurrentStatusName
      , f.FormStatusId
      , f.CurrentAssignedToUserName
      , f.IsRepeatableUsed
      , f.ReviewedLocal
      , f.IsTransferUsed
  from delta.`/mnt/silver/cab/form_headers` f
    left join delta.`/mnt/silver/cab/companies` c on c.CompanyId = f.CompanyId and c.InstanceId = f.InstanceId
```
