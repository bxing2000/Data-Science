```
companiesAttBronze = spark.read.format('parquet').load("/mnt/bronze/cab/operational/att/companies_client/")
companies = spark.read.format('delta').load("/mnt/silver/cab/companies")
accounts = spark.read.format('delta').load("/mnt/silver/cab/accounts")

companiesAttBronze.registerTempTable("companiesAttBronze")
companies.registerTempTable("companies")
accounts.registerTempTable("accounts")
  
```

```
%sql
select count(*)
from companies s
where s.InstanceId = 50 /*att*/
  and UserIntegrationType is null
```

```
%sql
select count(*)
from companies s
where s.InstanceId = 50 /*att*/
  and not exists (
    select *
    from companiesAttBronze b
    where b.CompanyId = s.CompanyId
  )
```
