```
from pyspark.sql.functions import year, month, dayofmonth, hour, col, lit, desc, dayofweek, concat, to_date, date_trunc
```

```
FormHeadersAtt = spark.read.format('parquet').load("/mnt/bronze/cab/operational/att/form_headers_client").withColumn('Instance', lit('ATT'))
FormHeadersEncore = spark.read.format('parquet').load("/mnt/bronze/cab/operational/encore/form_headers_client").withColumn('Instance', lit('Encore'))
FormHeadersVfIndia = spark.read.format('parquet').load("/mnt/bronze/cab/operational/vfindia/form_headers_client").withColumn('Instance', lit('VFIndia'))
FormHeadersMaxis = spark.read.format('parquet').load("/mnt/bronze/cab/operational/maxis/form_headers_client").withColumn('Instance', lit('Maxis'))
```

```
FormHeaders = FormHeadersAtt.union(FormHeadersEncore).union(FormHeadersVfIndia).union(FormHeadersMaxis)
```

```
totalCount = FormHeaders.count()
attCount = FormHeadersAtt.count()
print('ATT FormHeaders: ' + str(attCount))
print('Total FormHeaders: ' + str(totalCount))
print('% ATT: ' + str(attCount/totalCount))
```

### Note: Filter by year, month, day is not really accurate, it is using partitions which is first date of data loaded not all data, This will be fixed in silver zone

```
#First dates in warehouse
display(FormHeaders.filter(col('year') == 2016).sort(col('CreatedUtc')))
```

```
FormHeaders.filter(col('year') == 2019).count()
```

```
#newest dates in warehouse
display(FormHeaders.filter(col('year') == 2020).sort(col('CreatedUtc').desc()))
```

```
FormHeaders.filter((col('year') == 2020) & (col('month') == 4)).count()
```

```
FormHeaders.filter((col('year') == 2020) & (col('month') == 5) & (col('day') == 5)).count()
```

```
#Forms per day by instance
display(FormHeaders.filter(col('StartLocalTimeTag') > '2020-01-01')
        .groupby(col('Instance'), year(col('StartLocalTimeTag')), month(col('StartLocalTimeTag')), dayofmonth(col('StartLocalTimeTag'))).count()
        .groupby('Instance').agg({'count':'avg'}))
```

```
display(FormHeaders.filter((col('StartLocalTimeTag') > '2019-01-01') & (col('StartLocalTimeTag') < '2021-01-01'))
        #.filter((dayofweek(col('StartLocalTimeTag')) != 7) & (dayofweek(col('StartLocalTimeTag')) != 1))
        .groupby(col('Instance'), date_trunc('mon', 'StartLocalTimeTag')).count()
       )
```

```
#Only weekdays
display(FormHeaders.filter((col('StartLocalTimeTag') > '2019-01-01') & (col('StartLocalTimeTag') < '2021-01-01'))
        .filter((dayofweek(col('StartLocalTimeTag')) != 7) & (dayofweek(col('StartLocalTimeTag')) != 1))
        .groupby(col('Instance'), year(col('StartLocalTimeTag')), month(col('StartLocalTimeTag')), dayofmonth(col('StartLocalTimeTag'))).count()
        .groupby('Instance', col('year(StartLocalTimeTag)').alias('year'), col('month(StartLocalTimeTag)').alias('month')).agg({'count':'avg'})
        .withColumn('date', to_date(concat(col('year'), lit('-'), col('month'))))
       )
```
