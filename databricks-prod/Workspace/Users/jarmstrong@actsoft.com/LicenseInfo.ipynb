```
from pyspark.sql.functions import year, month, dayofmonth, hour, col, lit, desc, coalesce, udf,when
from pyspark.sql.types import IntegerType
```

```
Licenses = spark.read.format('parquet').load("/mnt/bronze/cab/billing/licenses")
```

```
display(Licenses)
```

```
LicencesWOriginalLicense = Licenses.filter(Licenses.OriginalLicenseId.isNotNull()) \
                                    .sort(col('licenseId')) \
                                   .groupby(col("OriginalLicenseId").alias('SwappedLicenseId')) \
                                    .agg({
                                      "Count": "count", 
                                      "OriginalActivationDate": "min", 
                                      "DeactDate": "max", 
                                      "Status": 'last'
                                    })
```

```
display(LicencesWOriginalLicense.sort(desc('count(Count)')))
```

```
BaseLicences = Licenses.filter(Licenses.OriginalLicenseId.isNull())
FilteredLicenses = BaseLicences.join(LicencesWOriginalLicense, col('LicenseId') == col('SwappedLicenseId'), how="left")
display(FilteredLicenses.filter(col('SwappedLicenseId').isNotNull()) )
```
