```
from pyspark.sql.functions import year, month, dayofmonth, hour, col, lit, desc
```

```
TripsAtt = spark.read.format('parquet').load("/mnt/bronze/cab/operational/att/trips_client").withColumn('Instance', lit('ATT'))
TripsEncore = spark.read.format('parquet').load("/mnt/bronze/cab/operational/encore/trips_client").withColumn('Instance', lit('Encore'))
TripsVfIndia = spark.read.format('parquet').load("/mnt/bronze/cab/operational/vfindia/trips_client").withColumn('Instance', lit('VFIndia'))
TripsMaxis = spark.read.format('parquet').load("/mnt/bronze/cab/operational/maxis/trips_client").withColumn('Instance', lit('Maxis'))
```

```
Trips = TripsAtt.union(TripsEncore).union(TripsVfIndia).union(TripsMaxis)
```

```
totalCount = Trips.count()
attCount = TripsAtt.count()
print('ATT Trips: ' + str(attCount))
print('Total Trips: ' + str(totalCount))
print('% ATT: ' + str(attCount/totalCount))
```

```
display(Trips.filter(col('year') == 2020).sort(col('StartUtcTime').desc()))
```

```
Trips.filter((col('year') == 2020) & (col('month') == 4)).count()
```

```
Trips.filter((col('year') == 2020) & (col('month') == 5) & (col('day') == 5)).count()
```

```
Trips.filter((col('year') == 2020) & (col('DistanceMiles') > 0)).agg({'DistanceMiles': 'sum'}).explain()
```

```
display(Trips.filter((col('year') == 2020) & (col('DistanceMiles') > 0)).agg({'DistanceMiles': 'sum'}))
```

```
display(Trips.filter((col('year') == 2020) & (col('month') == 4) ).agg({'IdleTimeSeconds': 'sum'}))
```

```
print(Trips.filter((col('year') == 2020) & (col('IdleTimeSeconds') > 1)).count())
print(Trips.filter((col('year') == 2020) & (col('IdleTimeSeconds') == 0)).count())
```

```
display(Trips.filter((col('year') == 2020) & (col('StopTimeSeconds') < 86400)).agg({'StopTimeSeconds': 'avg'}))
```
