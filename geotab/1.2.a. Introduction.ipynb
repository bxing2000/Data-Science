#Connecting to BQ (BigQuery)
These steps will allow you to authenticate to BQ and then run a query from within Colab


```
#@title Authenticate
from google.colab import auth
_ = auth.authenticate_user()
```

```
#@title Google Big Query Client 
from google.cloud import bigquery
client = bigquery.Client(project='geotab-bootcamp')
```

#Running queries in BQ  
 We will run a sample query and save it to a dataframe

```
# 1. we first save the SQL as a String of characters
SQL = ''' 
              select *
              from `geotab-bootcamp.Dashboarding.Trips`
              limit 10
'''
# 2. then we will run SQL and save it in a Pandas dataframe
df = client.query(SQL).to_dataframe()
df.head(2)
```
