### Group by Env, Year and Month and get he number of entries for each grouping

```
entriesPerMonth = testDF.groupby('env', 'year', 'month').count().orderBy('env', 'year', 'month' )
```

```
display(entriesPerMonth)
```

### Filter to only Encore Trips, then group by Env, Year and Month

```
filteredDF = testDF.filter(testDF.Env == "Encore")
#filteredDF = testDF.filter((testDF.Env == "Encore") & (testDF.month > 9))
display(filteredDF.groupby('env', 'year', 'month').count().orderBy('env', 'year', 'month'))
```

### Remove "deleted" trips then summarize all the distance column to get the total miles driven. Then format the number to make easier to read

```
from pyspark.sql.functions import format_number

totalMilesDriven = testDF.filter(testDF.Deleted == 'false').agg({"DistanceMiles":"sum"}).select(format_number('sum(DistanceMiles)', 0))
display(totalMilesDriven)
```

### Same as above but, this time group by enviroment first
Note: did not display here only stored table as new variable

```
totalDistanceByEnv = testDF.filter(testDF.Deleted == 'false').groupBy('env').sum('DistanceMiles').select('env', format_number('sum(DistanceMiles)', 0))
```

```
display(totalDistanceByEnv)
```

### Get the "Query Plan" of the grouping by env

```
totalDistanceByEnv.explain()
```

```
display(totalDistanceByEnv)
```

```
from pyspark.sql.functions import  col, min, max, avg, sum

avgDistanceByEnv = testDF.filter(testDF.Deleted == 'false') \
                         .groupBy('env') \
                         .agg( 
                              avg(col('DistanceMiles')).alias('avgDistance'), 
                              avg(col('StopTimeSeconds')).alias('avgStopTimeSec'), 
                              avg(col('IdleTimeSeconds')).alias('avgIdleTimeSec'), 
                              avg(col('TimeSinceLastTripSeconds')).alias('avgLastTripSec'), 
                              max(col('DistanceMiles')).alias('maxDistance'), 
                              max(col('StopTimeSeconds')/60).alias('maxStopTimeMin'), 
                              max(col('IdleTimeSeconds')/60).alias('maxIdleTimeMin'), 
                              max(col('TimeSinceLastTripSeconds')/60).alias('maxLastTripMin') 
                         )

display(avgDistanceByEnv)


val = [0, 1, 2, 3]
sub = [1:]
```

## join example

```
orderHeadersDF = spark.read.format('delta').load("/mnt/SandboxStorage/SampleDataSets/OrderHeaders_full_2019")
orderStatusesDF = spark.read.format('delta').load("/mnt/SandboxStorage/SampleDataSets/OrderStatuses_full_2019")
```

```
display(orderHeadersDF)
```

```
display(orderStatusesDF)
```

```
#inner join
joinedDF = orderHeadersDF.join(orderStatusesDF, orderHeadersDF.OrderId == orderStatusesDF.OrderId)
```

```
display(joinedDF.select('OrderNumberInt', 'StartUtcTime', 'StatusType', 'StartLat', 'StartLon'))
```
