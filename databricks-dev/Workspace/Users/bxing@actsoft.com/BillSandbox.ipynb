```
from pyspark.sql.functions import col, asc, countDistinct, current_date

devices = company_all_df.selectExpr("CompanyId", "Date(StartUtcTime) AS UsageDate", "DeviceId", "DistanceMiles").groupBy(["CompanyId", "UsageDate", "DeviceId"])
deviceCnt = devices.agg(countDistinct(col("DeviceId"))).alias("DeviceCount")

# deviceCnt = company_all_df.selectExpr("CompanyId", "Date(StartUtcTime) AS UsageDate", "DeviceId", "DistanceMiles").groupBy(["CompanyId", "UsageDate", "DeviceId"]).agg(countDistinct(col("DeviceId"))).alias("DeviceCount")

display(deviceCnt)
```
