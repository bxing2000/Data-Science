```
configs = { 
"fs.azure.account.auth.type": "CustomAccessToken",
"fs.azure.account.custom.token.provider.class": spark.conf.get("spark.databricks.passthrough.adls.gen2.tokenProviderClassName")
}
```

```
filesystem = 'sandboxes'
lakeName = 'devactsoftdatalake'
subDirectory = ''
mntName = 'SandboxStorage'
```

```
def iterator(x):
  return x.mountPoint == '/mnt/%s' % (mntName) 
def mount():
  dbutils.fs.mount(
    source = "abfss://%s@%s.dfs.core.windows.net/%s" % (filesystem, lakeName, subDirectory),
    mount_point = "/mnt/%s" % (mntName),
    extra_configs = configs)
  print('Mounted /mnt/%s' % (mntName))

result = list(filter(iterator, dbutils.fs.mounts()))

if (len(result) == 1):
  print('Sandboxes already mounted at /mnt/%s' % (mntName))
else:
  mount()
```
